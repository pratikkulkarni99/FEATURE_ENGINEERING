### Probobality ratio encoding

1.PROB OF SURVIVED BASED ON CABIN---CATEGORICAL VALUES

2.PROB OF NOT SURVIVED=1-PROB{SURVIVED}

3.DIVIDE PROB(SURV)/PROB(DIED)

4.DICTIONARY TO MAP CABIN WITH PROBABAILITY

5.REPLACE WITH CATEGORICAL FEATURES

```python
df=pd.read_csv("C:\\Users\\Pratik1\Desktop\dataset\\data's\\train.csv",usecols=['Cabin','Survived'])  #titanic dataset
df.head()
```

```python
df['Cabin'].fillna('missing',inplace=True) #replace with missing category
df.head()
```

```python
df['Cabin'].unique() #check unque categories in cabin
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]   #convert to string 
df.head()
```

```python
df['Cabin'].unique()  #unique
```

```python
prob_df=df.groupby(['Cabin'])['Survived'].mean()  #mean of cabin w.r.t survived
prob_df
```

```python
prob_df=pd.DataFrame(prob_df)
prob_df
```

```python
#calculate % person has died
prob_df['Died']=1-prob_df['Survived']
```

```python
prob_df.head()
```

```python
prob_df['Probability_ratio']=prob_df['Survived']/prob_df['Died']  #dividing 
prob_df.head()
```

```python
Probability_encoding=prob_df["Probability_ratio"].to_dict() #convert to dictionary
prob_df
```

```python
df['Cabin_encoded']=df['Cabin'].map(Probability_encoding) #mapping to ds
```
